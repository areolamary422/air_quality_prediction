```python
import pandas as pd                       # For data manipulation and CSV handling
from sklearn.model_selection import train_test_split  # For splitting data into training and testing sets
from sklearn.preprocessing import OneHotEncoder       # For converting categorical data (e.g., location) to numeric
from sklearn.linear_model import LinearRegression     # For building a simple regression model
from sklearn.metrics import mean_squared_error, r2_score  # For evaluating model performance
```

```python
import pandas as pd
df = pd.read_csv("C:/Users/USER/Downloads/air_quality_dataset.csv")
df.head()
```

```python
# Convert the categorical column 'location' into numeric values using One-Hot Encoding
# This creates new binary columns like location_Abuja, location_Kano (Lagos is dropped to avoid redundancy)
df_encoded = pd.get_dummies(df, columns=['location'], drop_first=True)

# Drop the 'date' column as it's not useful for prediction in this case
# Separate features (X) from the target (y)

X = df_encoded.drop(columns=['date', 'air_quality_index'])  # Features used for prediction
y = df_encoded['air_quality_index']
```

```python
# Split the dataset into 70% training and 30% testing data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)

```

```python
# Use the trained model to make predictions on the test data
y_pred = model.predict(X_test)
```

```python
# Print the Mean Squared Error (MSE) – lower is better
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))

# Print the R-squared score – closer to 1 means better accuracy
print("R^2 Score:", r2_score(y_test, y_pred))

# Optional: Show a table comparing actual and predicted values
pred_df = pd.DataFrame({'Actual': y_test.values, 'Predicted': y_pred})
print(pred_df.head())
```
